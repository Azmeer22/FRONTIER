```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

# -------------------- Data Loading --------------------

# File paths
file_paths = {
    "allergies": r"C:\Users\PCS\Desktop\allergies.csv",
    "careplans": r"C:\Users\PCS\Desktop\careplans.csv",
    "claims": r"C:\Users\PCS\Desktop\claims.csv",
    "claims_transactions": r"C:\Users\PCS\Desktop\claims_transactions.csv",
    "conditions": r"C:\Users\PCS\Desktop\conditions.csv",
    "devices": r"C:\Users\PCS\Desktop\devices.csv",
    "encounters": r"C:\Users\PCS\Desktop\encounters.csv",
    "imaging_studies": r"C:\Users\PCS\Desktop\imaging_studies.csv",
    "immunizations": r"C:\Users\PCS\Desktop\immunizations.csv",
    "medications": r"C:\Users\PCS\Desktop\medications.csv",
    "observations": r"C:\Users\PCS\Desktop\observations.csv",
    "organizations": r"C:\Users\PCS\Desktop\organizations.csv",
    "patients": r"C:\Users\PCS\Desktop\patients.csv",
    "payer_transitions": r"C:\Users\PCS\Desktop\payer_transitions.csv",
    "payers": r"C:\Users\PCS\Desktop\payers.csv",
    "procedures": r"C:\Users\PCS\Desktop\procedures.csv",
    "providers": r"C:\Users\PCS\Desktop\providers.csv",
    "supplies": r"C:\Users\PCS\Desktop\supplies.csv",
}

# Load datasets
datasets = {name: pd.read_csv(path) for name, path in file_paths.items()}

# Extract specific datasets
patients = datasets["patients"]
conditions = datasets["conditions"]
observations = datasets["observations"]
medications = datasets["medications"]
encounters = datasets["encounters"]

```

```python
# -------------------- Data Preprocessing --------------------

# Age calculation and grouping
patients['BIRTHDATE'] = pd.to_datetime(patients['BIRTHDATE'])
patients['AGE'] = (pd.Timestamp.now() - patients['BIRTHDATE']).dt.days // 365

def age_group(age):
    if age <= 10:
        return '0-10'
    elif age <= 18:
        return '11-18'
    elif age <= 64:
        return '19-64'
    else:
        return '65+'

patients['AGE_GROUP'] = patients['AGE'].apply(age_group)

# Psychiatric Diagnoses
psychiatric_conditions = ['depression', 'anxiety', 'schizophrenia']
conditions['DESCRIPTION'] = conditions['DESCRIPTION'].str.lower()
psychiatric_data = conditions[conditions['DESCRIPTION'].str.contains('|'.join(psychiatric_conditions))]

# PHQ-9 Scores
observations['DESCRIPTION'] = observations['DESCRIPTION'].str.lower()
phq9_data = observations[observations['DESCRIPTION'].str.contains('phq-9')]
phq9_data['VALUE'] = pd.to_numeric(phq9_data['VALUE'], errors='coerce')
phq9_data = phq9_data[phq9_data['VALUE'] > 10]

# Depression Treatment Analysis
depression_data = conditions[conditions['DESCRIPTION'].str.contains('depression', case=False)]
depression_encounters = pd.merge(
    depression_data,
    encounters.rename(columns={"Id": "ENCOUNTER"}),
    on="ENCOUNTER",
    how="left"
)
depression_encounters = depression_encounters.rename(columns={"PATIENT_x": "PATIENT"})
depression_treatment = pd.merge(depression_encounters, medications, on="PATIENT", how="left")
depression_treatment['DAYS_TO_TREATMENT'] = (
    pd.to_datetime(depression_treatment['START_x']) - pd.to_datetime(depression_treatment['START_y'])
).dt.days
untreated_depression = depression_treatment[depression_treatment['DAYS_TO_TREATMENT'].isnull()]

# CPT Code Utilization
psychiatric_cpt = encounters[encounters['DESCRIPTION'].str.contains('|'.join(psychiatric_conditions), case=False)]
cpt_code_counts = psychiatric_cpt['CODE'].value_counts()
```

```python
# -------------------- Visualizations --------------------

# Age Group Distribution
age_group_counts = patients['AGE_GROUP'].value_counts()
plt.figure(figsize=(8, 6))
age_group_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Age Group Distribution')
plt.xlabel('Age Group')
plt.ylabel('Number of Patients')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('age_group_distribution.png')
plt.show()

# Psychiatric Diagnoses
psychiatric_counts = psychiatric_data['DESCRIPTION'].value_counts()
plt.figure(figsize=(8, 6))
psychiatric_counts.plot(kind='bar', color='coral', edgecolor='black')
plt.title('Psychiatric Diagnoses Count')
plt.xlabel('Condition')
plt.ylabel('Number of Patients')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('psychiatric_diagnoses.png')
plt.show()

# PHQ-9 Scores by Diagnosis
phq9_by_diagnosis = pd.merge(phq9_data, conditions, on='PATIENT', how='left')
phq9_by_diagnosis_counts = phq9_by_diagnosis['DESCRIPTION_y'].value_counts()
plt.figure(figsize=(10, 6))
phq9_by_diagnosis_counts.plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title('Patients with PHQ-9 > 10 by Diagnosis')
plt.xlabel('Diagnosis')
plt.ylabel('Number of Patients')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('phq9_scores.png')
plt.show()

# CPT Code Utilization
cpt_code_counts_df = cpt_code_counts.reset_index()
cpt_code_counts_df.columns = ['CPT Code', 'Frequency']  # Explicitly set column names

fig = px.bar(
    cpt_code_counts_df,
    x='CPT Code',
    y='Frequency',
    title='CPT Code Utilization for Psychiatric Diagnoses',
    labels={'CPT Code': 'CPT Code', 'Frequency': 'Frequency'},
    color='Frequency',
    color_continuous_scale='Bluered'
)
fig.write_image('cpt_code_utilization.png')
fig.show()
```

```python
# Insights Summary
insights = {
    "Age Group Distribution": age_group_counts.to_dict(),
    "Psychiatric Diagnoses": psychiatric_counts.to_dict(),
    "PHQ-9 Scores > 10": phq9_by_diagnosis_counts.to_dict(),
    "Untreated Depression Patients": untreated_depression['PATIENT'].nunique()
}

print("Insights Summary:")
for key, value in insights.items():
    print(f"{key}: {value}")
```
